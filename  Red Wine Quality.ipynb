```python
#Importing of the libraries in the dataset
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
```

```python
#Loading of the dataset
path='/content/drive/MyDrive/Dataset/winequality-red.csv'
df=pd.read_csv(path)
df
```

```python
df.columns #Checking of the columns
```

```python
df.isnull().sum() #Finding of the missing values
```

```python
df.describe()
```

```python
duplicates = df[df.duplicated()] #Checking for duplicates
dff = df.drop_duplicates() #Dropping of duplicates
```

```python
dff
```

```python
dff.nunique() #Unique values of dataset
```

```python
#Histogram of Alcohol Density
sns.histplot(dff['alcohol'],color='red');
plt.xlabel("Alcohol")
plt.ylabel("Density")
plt.title("Histogram of Alcohol Density");
```

```python
#Fixed Acidity Vs Volatile Acidity
sns.scatterplot(x='fixed acidity', y='volatile acidity', hue='quality',data=dff);
plt.xlabel("Fixed Acidity")
plt.ylabel("Volatile Acidity")
plt.title("Fixed Acidity vs Volatile Acidity");
```

```python
#Relation of Alcohol with Quality
sns.barplot(x='quality', y='alcohol', color = 'skyblue',data=dff)
plt.title('Relation of alcohol with Wine')
plt.xlabel('Quality')
plt.ylabel('Alcohol')
plt.show()
```

```python
plt.figure(figsize=(18,12))
#Quality Vs Citric Acid"
plt.subplot(2,2,1) #2-Rows, 2-column, 1-first
sns.barplot(x = 'quality', y = 'citric acid',color='green', data = dff)
plt.title("Quality Vs Citric Acid")
plt.xlabel("Quality")
plt.ylabel("Citric Acid")

#Quality Vs Chlorides
plt.subplot(2, 2, 2)
sns.scatterplot(x = 'quality', y = 'chlorides',color='orange',data = dff)
plt.title("Quality Vs Chlorides")
plt.xlabel("Quality")
plt.ylabel("Chlorides")

#Quality Vs Free Sulphurdioxide
plt.subplot(2,2,3)
sns.violinplot(x = 'quality', y = 'free sulfur dioxide',color='purple', data = dff)
plt.title("Quality Vs Free Sulphur dioxide")
plt.xlabel("Quality")
plt.ylabel("Free Sulphate dioxide")

#Quality Vs Sulphates
plt.subplot(2,2,4)
sns.barplot(x = 'quality', y = 'sulphates',color='#FF6347', data = dff)
plt.title("Quality Vs Sulphates")
plt.xlabel("Quality")
plt.ylabel("Sulphates")
```

```python
#Violin Plot of Wine Quality by pH Levels
sns.set_style('darkgrid')
plt.figure(figsize=(8, 6))
sns.violinplot(x='quality', y='pH', data=dff, palette='deep')
plt.title('Violin Plot of Wine Quality by pH Levels')
plt.xlabel('Quality')
plt.ylabel('pH')
plt.show()
```

```python
#Swarm Plot of Wine Quality by Density
sns.set_style('darkgrid')
plt.figure(figsize=(8, 6))
sns.swarmplot(x='quality', y='density', data=dff, palette='dark')
plt.title('Swarm Plot of Wine Quality by Density')
plt.xlabel('Quality')
plt.ylabel('Density')
plt.show();
```

```python
#Distribution of Fixed Acidity, Volatile Acidity, Citric Acid
plt.figure(figsize=(40,12))

plt.subplot(1, 3, 1)
sns.histplot(x=dff["fixed acidity"], bins=15, kde=True ,color='#29066B')
plt.xlabel("Fixed Acidity")
plt.ylabel("Density")
plt.title("Distribution of Fixed Acidity");

plt.subplot(1, 3, 2)
sns.histplot(x=dff["volatile acidity"], bins=15,kde=True ,color='#29066B')
plt.xlabel("Volatile Acidity")
plt.ylabel("Density")
plt.title("Distribution of  Fixed Acidity");

plt.subplot(1, 3, 3)
sns.histplot(x=dff["citric acid"],bins=15, kde=True ,color='#29066B')
plt.xlabel("Citric Acid")
plt.ylabel("Density")
plt.title("Distribution of Citric Acid");
plt.tight_layout()
plt.show();
```

```python
#Distribution of Residual sugar,Chlorides,Free sulfurdioxide
plt.figure(figsize=(30,12))

plt.subplot(1, 3, 1)
sns.histplot(x=dff["residual sugar"], bins=25,kde=True ,color='red')
plt.xlabel("Residual sugar")
plt.ylabel("Density")
plt.title("Distribution of Residual sugar");

plt.subplot(1, 3, 2)
sns.histplot(x=dff["chlorides"],bins=20, kde=True ,color='red')
plt.xlabel("Chloride")
plt.ylabel("Density")
plt.title("Distribution of Chloride")

plt.subplot(1, 3, 3)
sns.histplot(x=dff["free sulfur dioxide"],bins=20, kde=True ,color='red')
plt.xlabel("Free Sulfurdioxide")
plt.ylabel("Density")
plt.title("Distribution of Free Sulfurdioxide");
plt.tight_layout()
plt.show();

```

```python
#Distribution of Sulphates
sns.histplot(x=dff["sulphates"],bins=20, kde=True ,color='blue')
plt.xlabel("Sulphates")
plt.ylabel("Density")
plt.title("Distribution of Sulphates");
```

```python
#Correlation Matrix
plt.figure(figsize=(40,10))
plt.title('Correlation Matrix')
sns.heatmap(dff.corr(), annot=True,cmap='Blues')
plt.show()
```
